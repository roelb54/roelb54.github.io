### lists, tuples, matrices

```python
import numpy as np

x = np.array([2, 7, 5])  # list: mutable, ordered
x
```

```python
import numpy as np

y = np.linspace(4, 10, 3)
y
```

```python
print(
    x + y,
    x / y,
    x ** y)
```

```python
print(
    x[1],      # second element
    x[[1]],    # slice: only second element (returned as list)
    x[1:3],    # slice: second and third element 
    x[:2],     # slice: first 2 elements
    x[-2:],    # slice: last two elements
    x[[0, -1]] # slice: first and last element
)
```

### random data, graphics

```python
x = np.random.uniform(size=50)
x = np.random.uniform(low=0, high=1, size=(10,5))
x = np.random.uniform(low=0, high=1, size=50)
x
```

```python
y = np.random.normal(size=50)
y = np.random.normal(loc=0, scale=1, size=(10,5))
y = np.random.normal(loc=0, scale=1, size=50)

y
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns

sns.set()

plt.scatter(x,y)
plt.show()
```

```python
sns.scatterplot(x=x, y=y)
plt.show()
```

```python
fig, axs = plt.subplots(nrows=2,ncols=1)

axs[0].scatter(x,y)
axs[1].hist(y)

plt.show()
```

```python
fig, axs = plt.subplots(nrows=2,ncols=1)

fig.suptitle('multiple plots')

sns.scatterplot(x=x,y=y, ax=axs[0])
sns.histplot(y, ax=axs[1], stat='frequency')

plt.show()
```
