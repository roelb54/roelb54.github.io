```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
t = np.linspace(0, 2, 21)
y = np.linspace(0, 2, 21)

X, Y = np.meshgrid(t, y)

dy = lambda k :  -k * Y
dt = np.ones(Y.shape)

plt.quiver(X, Y, dt, dy(1), color='k')
plt.title('Flowfield $\\frac{dy(t)}{dt}=-k \cdot y(t)$')
plt.show()
```
